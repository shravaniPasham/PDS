```python
#importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the cleaned data
cleaned_data = pd.read_csv("/content/cleaned_data.csv")
```

```python
#Histogram of Weight
plt.figure(figsize=(8, 6))
sns.histplot(cleaned_data['Weight'], bins=10, kde=True)
plt.title("Histogram of Weight")
plt.savefig("histogram_Weight.png")
plt.show()
```

```python
#Scatter plot of Grip Strength vs Weight
plt.figure(figsize=(6, 6))
sns.scatterplot(x='Weight', y='grip_strength_kilograms', data=cleaned_data)
plt.title("Scatter plot of Grip strength vs. Weight")
plt.savefig("scatter_weight_grip.png")
plt.show()
```
