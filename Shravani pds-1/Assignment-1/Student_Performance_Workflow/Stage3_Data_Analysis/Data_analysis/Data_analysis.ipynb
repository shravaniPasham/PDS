```python
#importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the cleaned data
data = pd.read_csv("/content/cleaned_data.csv")
```

```python
#Histogram of Math Scores
plt.figure(figsize=(8, 6))
sns.histplot(data['math score'], bins=10, kde=True)
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.savefig('Distribution_of_Math_Scores.png')
plt.show()
```

```python
#Box Plot of Math Scores Grouped by Education Level
plt.figure(figsize=(10, 6))
sns.boxplot(x='parental level of education', y='math score', data=data)
plt.title('Math Scores by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Math Score')
plt.xticks(rotation=45)
plt.savefig('Math_Scores_by_Education_Level.png')
plt.show()
```

```python
#Pair Plot of Scores
sns.pairplot(data, vars=['math score', 'reading score', 'writing score'], hue='gender')
plt.suptitle('Pair Plot of Scores by Gender')
plt.savefig('Pair_plot_of_Scores_by_Gender.png')
plt.show()
```

```python
#Heatmap of Correlations
correlation_matrix = data[['math score', 'reading score', 'writing score']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Scores')
plt.savefig('Correlation_Heatmap_of_Scores.png')
plt.show()
```

```python
#  Box Plot of Reading Scores by Education Level
plt.figure(figsize=(10, 6))
sns.boxplot(x='parental level of education', y='reading score', data=data)
plt.title('Reading Scores by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Reading Score')
plt.xticks(rotation=45)
plt.savefig('Reading_Scores_by_Education_Level.png')
plt.show()
```
